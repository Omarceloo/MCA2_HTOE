## Actividad:
Problema Interprepas Matemáticas VI área 1

Sean las funciones
\begin{equation}
f(x) = \sqrt{-x} \qquad g(x) = \ln x
\end{equation}
gráfica
\begin{equation}
f\circ g \qquad g ∘ f \qquad fg
\end{equation}


```python
from sympy import symbols, sqrt, ln
from sympy.plotting import plot

x = symbols('x')

def f(x):
    return sqrt(-x)

def g(x):
    return ln(x)

fog = f(g(x))   # sqrt(-ln(x))
gof = g(f(x))   # ln(sqrt(-x))
fg  = f(x)*g(x) # sqrt(-x)*ln(x)

g1 = plot(fog, (x, 0.01, 1),     line_color='red',   label='f∘g', show=False)
g2 = plot(gof, (x, -10, 0),      line_color='blue',  label='g∘f', show=False)
g3 = plot(fg,  (x, -10, 0),      line_color='green', label='f·g', show=False)

g1.append(g2[0])
g1.append(g3[0])
g1.show()
```


## Actividad:

Consider the following function.

\begin{equation}
f(x) = \sum_{k=1}^{100} \left( \frac{\sin(2\pi k^2 x)}{4\pi^2k^5} + \frac{x^2}{2k}\right)
\end{equation}
This function was suggested by Sungkon Chang as an example of a function
that looks quite “innocent” but whose derivatives behave quite wildly. Plot 𝑓 (𝑥),
𝑓 ′ (𝑥) and 𝑓 ′′ (𝑥) and observe this behaviour.

```python
from sympy import symbols, sin, pi, summation, diff
from sympy.plotting import plot

x, k = symbols('x k')

f = summation(sin(2*pi*k*2*x)/(4*pi**2*k**5) + x**2/(2*k), (k, 1, 100))

df  = diff(f, x)
ddf = diff(f, x, 2)

g1 = plot(f,   (x, -1, 1), line_color='blue',  label='f(x)',   show=False)
g2 = plot(df,  (x, -1, 1), line_color='red',   label="f'(x)",  show=False)
g3 = plot(ddf, (x, -1, 1), line_color='green', label="f''(x)", show=False)

g1.append(g2[0])
g1.append(g3[0])
g1.show()
```

## Actividad:
Grafica las siguientes funciones implícitas

* 1)

\begin{equation}
x^2 + y^3 - 2y = 3
\end{equation}
* 2) Under certain conditions, van der Waals’ equation relating the pressure P and volume
V of a gas is


\begin{equation}
\left(P + \frac{5}{V^2} \right)(V - 0.03) = 9.7
\end{equation}

Referencia: Smith (2013), Calculus, 4th edition pag 156 y 157

```python
from sympy import symbols, Eq
from sympy import plot_implicit

x, y = symbols('x y')
P, V = symbols('P V')

# 1. x² + y³ - 2y = 3
g1 = plot_implicit(Eq(x**2 + y**3 - 2*y, 3), (x, -6, 6), (y, -4, 2.5),
                   title='x² + y³ - 2y = 3', show=False)

# 2. van der Waals: (P + 5/V²)(V - 0.03) = 9.7
g2 = plot_implicit(Eq((P + 5/V**2) * (V - 0.03), 9.7), (V, 0.1, 5), (P, -20, 10),
                   title='Van der Waals', show=False)

g1.show()
g2.show()
```

## Actividad:
Grafica las siguientes funciones paramétrizadas

* 1) With $0 \leq t \leq 14\pi$
\begin{align}
𝑥(𝑡) &= cos(𝑡) + 1/2 cos(7𝑡) + 1/3 sin(17𝑡) \\
𝑦(𝑡) &= sin(𝑡) + 1/2 sin(7𝑡) + 1/3 cos(17𝑡)
\end{align}

```python
from sympy import symbols, cos, sin, pi
from sympy.plotting import plot_parametric

t = symbols('t')

# Definición de las funciones paramétricas
x_t = cos(t) + (1/2)*cos(7*t) + (1/3)*sin(17*t)
y_t = sin(t) + (1/2)*sin(7*t) + (1/3)*cos(17*t)

#Se grafica en el intervalo dado
g1 = plot_parametric(x_t, y_t, (t, 0, 14*pi),
                     title='Gráfica Paramétrica',
                     xlabel='x(t)', ylabel='y(t)',
                     show=False)

g1.show()
```

Cambia el valor de parámetro *h* a:

* I) 1 , 2, 3, 5 , 6
* II)¿Qué notas en la gáficas?

```python
import numpy as np
import matplotlib.pyplot as plt

valores_h = [1, 2, 3,5,6]
fig, axes = plt.subplots(1, 5, figsize=(18, 4))
theta = np.linspace(0, 2 * np.pi, 4000)

for i, h in enumerate(valores_h):
    r = 3 * np.cos(h * theta)

    x = r * np.sin(theta)
    y = r * np.cos(theta)
    ax = axes[i]
    ax.plot(x,y,   color='darkblue',linewidth=1.5)
    ax.set_title(f"h = {h}", fontsize=11 , fontweight= 'bold')
    ax.axis('off')
    ax.axis('equal')
plt.tight_layout()
plt.show()
```

**¿Que conductas notas en la grafica dada?** <br>
Se puede notar claramente conforme se va aumentando h se va haciendo una flor diferente, cuando h es par la grafica tiene el doble de hojas,  pero cuando es impar solo se queda con ese número de hojas.
Como la función es simétrica, para h's impares se acaban de dibujar las hojas  al dar la vuelta de cero a pi, porque en la segunda pasada, de pi a dos pi se vuelven a dibujar esas mismas hojas de la primera vuelta, como si soloamente se hubiera hecho una vez el recorrido. Pero cuando las h son pares, la formula da valores negativos y en la segunda vuelta en lugar de dibujar otra vez las mismas hojas, dibuja el mismo numero de hojas pero en espacios vacíos en los lados opuestos.

## Lemniscata en coord. polares
$$
r^2 = a^2 \cos(2\theta)
$$
donde
$$
x = r\cos(\theta) \;\; \; \; y = r\sin(\theta)
$$



```python
import numpy as np
import matplotlib.pyplot as plt

valores_a = [1, 2, 3, 5, 6]
theta = np.linspace(0, 2*np.pi, 2000)

plt.figure(figsize=(7,7))

for a in valores_a:
    r_cuadrado = a**2 * np.cos(2*theta)
    r = np.sqrt(np.maximum(r_cuadrado, 0))  # evita números negativos
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.plot(x, y, label=f"a = {a}")

plt.title("Lemniscata de Bernoulli para distintos valores de a")
plt.axis('equal')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## Actividad:
Grafica las siguientes funciones dadas en coordenadas polares obteniendo las coordenadas x,y parametrizadas

* 1) Cardioide
\begin{align}
r(θ) &= a(1 + \cos(θ)) \\
\end{align}

### Polares. Directamente con la función matplotlib.pyplot.polar(...)

```python
import numpy as np
import matplotlib.pyplot as plt

a = 1
theta = np.linspace(0, 2 * np.pi, 1000)
r = a * (1 + np.cos(theta))

x = r * np.cos(theta)
y = r * np.sin(theta)
plt.figure(figsize=(6, 6))
plt.plot(x, y,    color='darkblue',linewidth=2)
plt.title(r"$r(\theta) = a(1 + \cos(\theta))$", fontsize=12, fontweight='bold')
plt.xlabel("x")
plt.ylabel("y")

plt.grid(True,    linestyle='--', alpha=0.5 )
plt.axis('equal')
plt.show()
```

* 1) Espiral Hiperbólica
